```python
import os
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.ndimage import zoom
from scipy.ndimage import gaussian_filter
import scipy
from scipy import ndimage
print ('modules loaded')
```

```python
class DataGenerator():
    def __init__(self, df, batch_size = 1):
        self.df = df
        self.batchsize = batch_size
    
    def gaussian_filter(self,image):
        for i in range(128):
            image[:,:,i] = gaussian_filter(image[:,:,i], sigma=1)

        return image

    def sharpening(self,image):
        for i in range(128):
            blurred_f = ndimage.gaussian_filter(image[:,:,i], 3)
            filter_blurred_f = ndimage.gaussian_filter(blurred_f, 1)
            alpha = 40
            image[:,:,i] = blurred_f + alpha * (blurred_f - filter_blurred_f)
        
        return image

    def high_pass(self,image):
        for i in range(128):
            lowpass = ndimage.gaussian_filter(image[:,:,i], 3)
            image[:,:,i] = image[:,:,i] - lowpass
        
        return image

    def generate(self):
        while True:
            batchcount = 0
            for index, row in self.df.iterrows():
                img = cv2.imread(row['filepaths'])
                img = cv2.resize(img, (224,224))
                img = img/255.0
                label = row['labels']
                batchcount = batchcount + 1
                if batchcount == self.batchsize:
                    if index == 0:
                        yield img, np.array([label])

                    elif index == 1:
                        res = self.gaussian_filter(img)
                        yield res,np.array([label])

                    elif index == 2:
                        res = self.sharpening(img)
                        yield res,np.array([label])
                    
                    elif index == 3:
                        res = self.high_pass(img)
                        yield res,np.array([label])
                    
                    batchcount = 0

    # def predict_generate(self):
    #         while True:
    #             batchcount = 0
    #             # for imagefile in self.imagefiles:
    #             #     image1 = np.expand_dims(self.resize(nib.load(imagefile[0])),axis=-1)
    #             #     image2 = np.expand_dims(self.resize(nib.load(imagefile[1])),axis=-1)
    #             #     image3 = np.expand_dims(self.resize(nib.load(imagefile[2])),axis=-1)
    #             #     image4 = np.expand_dims(self.resize(nib.load(imagefile[3])),axis=-1)
    #             #     image5 = np.expand_dims(self.resize(nib.load(imagefile[4])),axis=-1)

    #                 batchcount = batchcount + 1
    #                 if batchcount == self.batchsize:
    #                     img_list = np.array([image1,image2,image3,image4,image5])
    #                     yield img_list
    #                     batchcount = 0
        
```
